## 레이블 인코딩

카테고리 피처를 코드형 숫자 값으로 변환

```python
from sklearn.preprocessing import LabelEncoder
```

```python
items = ['Man','Woman','Nonbinary','Prefer not to say','Prefer to self-describe']
```

```python
# LabelEncoder를 객체로 생성한 후, fit()과 transform()으로 레이블 인코딩 수행
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
print('인코딩 변환값: ', labels)
print('인코딩 클래스: ', encoder.classes_)
```

## 원 핫 인코딩 (One Hot Encoding)

OneHotEncoder 클래스 통해 scikit-learn 라이브러리에서 사용
범주형 입력 변수를 인코딩

피러 값의 유형에 따라 새로운 피처를 추가하고 고유 값에 해당하는 칼럼에만 1을 표시하고 나머지 칼럼에는 0을 표시. 즉, 행 형태로 되어 있는피처의 고유 값을 열 형태로 차원을 변경한 뒤, 고유값에 해당하는 칼럼에만 1을 표시하고 나머지 칼럼에는 0을 표시

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
import numpy as np
```

```python
items = ['Man','Woman','Nonbinary','Prefer not to say','Prefer to self-describe']
```

```python
# 먼저 숫자 값으로 변환 위해 LabelEncoder로 변환
encoder = LabelEncoder()
encoder.fit(items)
labels = encoder.transform(items)
```

```python
# 2차원 데이터로 변환
labels = labels.reshape(-1, 1)
```

```python
# 원-핫 인코딩 적용
OH_encoder = OneHotEncoder()
OH_encoder.fit(labels)
OH_labels = OH_encoder.transform(labels)
```

```python
print('원-핫 인코딩 데이터')
OH_labels.toarray()
```

```python
print('원-핫 인코딩 데이터 차원')
labels.shape
```
